### Maximum Population

```python
sorted_df=df.sort_values(by='Population',ascending=False)
```

```python
plt.figure(figsize=(15,6))
colors = sns.color_palette('Set3', 10)
top=sorted_df.head(10)
plt.bar(top['Country'],top['Population'],color=colors,edgecolor='k')
plt.title('Population wise Top 10 Countries',size=15)
plt.xlabel('Countries')
plt.ylabel('Population in Billions')

plt.xticks(rotation=45)
    
plt.show()
```

### Maximum Population density

```python
plt.figure(figsize=(15,6))
top_dens=df.sort_values(by='Density\n(P/Km2)',ascending=False).head(10)
plt.bar(top_dens['Country'],top_dens['Density\n(P/Km2)'],edgecolor='k')
plt.title('Countries with Maximum Population Desnity')
plt.xticks(rotation=45)
plt.xlabel('Countries')
plt.ylabel('Population density in thousand')
plt.show()
```

### Correlation Between Tax Revenue and Total Tax Rate

```python
sns.regplot(x=df['Total tax rate'],y=df['Tax revenue (%)'])
plt.title("Correlation Between Tax Revenue and Total Tax Rate")
plt.show()
print('There is slightly negetive correlation between Tax rate and Tax Revenue ')
```

```python
numcol=df[['Density\n(P/Km2)','Agricultural Land( %)',
        'Land Area(Km2)','Birth Rate','Co2-Emissions','Fertility Rate',
        'Forested Area (%)','GDP','Gross primary education enrollment (%)',
        'Gross tertiary education enrollment (%)','Infant mortality','Life expectancy'
        ]]

plt.figure(figsize=(12,6))
Correlation_matrix=numcol.corr()
sns.heatmap(Correlation_matrix,annot=True,cmap='plasma',linewidths=.5)
plt.title('Correlation Matrix of Numeric Columns')
plt.xticks(size=7)
plt.yticks(size=7)
plt.show()

```

```python
plt.figure(figsize=(16, 12))
sns.clustermap(Correlation_matrix, annot=True, cmap='plasma', linewidths=0.5)
plt.title('Correlation Heatmap for Numeric Variables')
plt.show()
```
